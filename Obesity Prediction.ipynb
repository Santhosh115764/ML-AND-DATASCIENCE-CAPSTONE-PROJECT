```python
X=dataset.drop('NObeyesdad',axis=1)
Y=dataset['NObeyesdad']
```

```python
print(X.shape)
print(Y.shape)
```

```python
X.select_dtypes(include='object').columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in X.select_dtypes(include='object').columns:
    X[col] = le.fit_transform(X[col])
```

```python
Y=le.fit_transform(Y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y_test,Y_pred)
```

```python
Y_test
```

```python
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt

plt.imshow(cm, cmap='Blues')
plt.colorbar()
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dataset_model=DecisionTreeClassifier(random_state=42)
dataset_model.fit(X_train,Y_train)
```

```python
dataset_Y_pred=dataset_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
dataset_accuracy=accuracy_score(Y_test,dataset_Y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt

plt.imshow(cm, cmap='Greens')
plt.colorbar()
plt.title("Decision Tree Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print("Logistic Regression Accuracy:", accuracy)
print("Decision Tree Accuracy:", dataset_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model= RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(X_train, Y_train)
```

```python
rf_y_pred = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
rf_accuracy = accuracy_score(Y_test, rf_y_pred)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, rf_y_pred))
```

```python
from sklearn.metrics import confusion_matrix
rf_cm = confusion_matrix(Y_test, rf_y_pred)
print(rf_cm)
```

```python
import matplotlib.pyplot as plt

plt.imshow(rf_cm, cmap='Oranges')
plt.colorbar()
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print("Logistic Regression :", accuracy)
print("Decision Tree       :", dataset_accuracy)
print("Random Forest       :", rf_accuracy)
```
